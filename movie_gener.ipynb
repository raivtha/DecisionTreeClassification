```python
# Generate some random data with outliers
data = np.concatenate([np.random.normal(0, 1, 50), np.random.normal(10, 1, 10)])
```

```python
# Plot a boxplot of the original data
plt.boxplot(data)
plt.title('Original data')
plt.show()
```

```python
# Calculate the interquartile range (IQR) and set a threshold for outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
outlier_threshold = Q3 + 1.5*IQR
```

```python

# Remove outliers from the data
data_no_outliers = data[data <= outlier_threshold]
```

```python
# Plot a boxplot of the data without outliers
plt.boxplot(data_no_outliers)
plt.title('Data without outliers')
plt.show()
```

## Feature extraction with pca

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA()
pca.fit(X_train_N)
```

```python
plt.figure(1, figsize=(14, 13))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_ratio_, linewidth=2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_')
```
